# Model Evaluation & Quality Metrics

This notebook demonstrates how we evaluate the prediction accuracy and recommendation ranking quality of our Collaborative Filtering, SVD, and Hybrid systems on the testing partition.

```python
import os
import sys
import pandas as pd
import numpy as np

# Align imports
sys.path.append(os.path.dirname(os.getcwd()))
import config
from src.utils import load_pkl
from src.recommendation import HybridRecommender
from src.evaluation import evaluate_predictions, evaluate_precision_recall_at_k
```

## 1. Load Pretrained Models & Test Set

```python
data = load_pkl(config.ENCODED_DATA_PKL)
test_ratings = data['test_ratings']
recommender = HybridRecommender()
print(f"Testing dataset samples: {len(test_ratings)}")
```

## 2. Calculate Evaluation Comparison

```python
methods = ['cf', 'svd', 'hybrid']
for m in methods:
    rmse, mae = evaluate_predictions(recommender, test_ratings, method=m)
    precision, recall, f1 = evaluate_precision_recall_at_k(recommender, test_ratings, k=10, method=m)
    print(f"[{m.upper()} Model] RMSE: {rmse:.4f} | MAE: {mae:.4f} | Precision@10: {precision:.4f} | Recall@10: {recall:.4f} | F1: {f1:.4f}")
```
