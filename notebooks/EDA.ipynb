# Exploratory Data Analysis (EDA) - Movie Recommendation System

This notebook demonstrates exploratory analysis on our synthetic movie ratings dataset. We will load the generated raw data (`movies.csv`, `users.csv`, `ratings.csv`), compute system statistics, explore rating distributions, and analyze movie popularity.

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure config can be loaded
sys.path.append(os.path.dirname(os.getcwd()))
import config
```

## 1. Load Datasets

```python
movies = pd.read_csv(config.MOVIES_CSV)
users = pd.read_csv(config.USERS_CSV)
ratings = pd.read_csv(config.RATINGS_CSV)

print(f"Movies Catalog Count: {len(movies)}")
print(f"Users Catalog Count: {len(users)}")
print(f"Ratings Log count: {len(ratings)}")
```

## 2. Ratings Sparsity Analysis

```python
n_users = ratings['user_id'].nunique()
n_movies = ratings['movie_id'].nunique()
n_possible = n_users * n_movies
n_actual = len(ratings)
sparsity = 1.0 - (n_actual / n_possible)

print(f"Unique Users in Ratings: {n_users}")
print(f"Unique Movies in Ratings: {n_movies}")
print(f"Rating Matrix Sparsity: {sparsity * 100:.2f}%")
```

## 3. Visualize Ratings Distribution

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='rating', data=ratings, palette='viridis')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating Value')
plt.ylabel('Count')
plt.show()
```

## 4. Popular Movie Genre Analysis

```python
# Split genres and count occurrence
genres_series = movies['genres'].str.split('|').explode()
plt.figure(figsize=(12, 6))
sns.countplot(y=genres_series, order=genres_series.value_counts().index, palette='magma')
plt.title('Movie Catalog Count by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```
