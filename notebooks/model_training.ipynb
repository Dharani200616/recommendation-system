# Model Training - Recommendation Engine

This notebook demonstrates how we train our **User-Based Collaborative Filtering** and **Matrix Factorization (SVD)** recommendation engines on preprocessed movie rating logs.

```python
import os
import sys
import pandas as pd
import numpy as np

# Ensure path alignment
sys.path.append(os.path.dirname(os.getcwd()))
import config
from src.preprocessing import preprocess_pipeline
from src.collaborative_filtering import UserCollaborativeFiltering
from src.matrix_factorization import SVDMatrixFactorization
```

## 1. Preprocess & Load train/test splits

```python
data = preprocess_pipeline()
train_ratings = data['train_ratings']
print(f"Training set samples: {len(train_ratings)}")
```

## 2. Train Collaborative Filtering Model

```python
cf = UserCollaborativeFiltering()
cf.fit(train_ratings)

# Inspect User similarity matrix shape
print("User Similarity Matrix shape:", cf.user_similarity.shape)
```

## 3. Train SVD Latent Matrix Factorization Model

```python
svd = SVDMatrixFactorization(n_factors=15, epochs=10, lr=0.005)
svd.fit(train_ratings, verbose=True)
```
